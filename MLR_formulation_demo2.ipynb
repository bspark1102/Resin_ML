# 1. Training of Surrogate Models

```python
import os
import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score, root_mean_squared_error
from scipy.optimize import differential_evolution

TRAIN_DIR = "train_data"

X_tr  = np.load(os.path.join(TRAIN_DIR, "X.npy")).astype(float)
yV_tr = np.load(os.path.join(TRAIN_DIR, "y_viscosity.npy")).astype(float)
yE_tr = np.load(os.path.join(TRAIN_DIR, "y_elongation.npy")).astype(float)
yM_tr = np.load(os.path.join(TRAIN_DIR, "y_modulus.npy")).astype(float)
yR_tr = np.load(os.path.join(TRAIN_DIR, "y_reactiontime2.npy")).astype(float)

# to fraction
rs = X_tr.sum(axis=1)
if np.allclose(rs, 100.0, atol=1e-6):
    X_tr = X_tr / 10.0
elif np.allclose(rs, 1.0, atol=1e-6):
    X_tr = X_tr * 10.0

# normalization
yV_tr_log = np.log1p(yV_tr)
yE_tr_log = np.log1p(yE_tr)
yM_tr_log = np.log1p(yM_tr)
yR_tr_log = np.log1p(yR_tr)

visc_model = SVR(kernel='rbf', C=100,  gamma=0.1,  epsilon=0.01)
elon_model = SVR(kernel='rbf', C=400,  gamma=0.01, epsilon=0.01)
mod_model  = SVR(kernel='rbf', C=17, gamma=0.04,  epsilon=0.13)
rt_model   = SVR(kernel='rbf', C=68,  gamma=0.03,  epsilon=0.1)

visc_model.fit(X_tr, yV_tr_log)
elon_model.fit(X_tr, yE_tr_log)
mod_model.fit(X_tr, yM_tr_log)
rt_model.fit(X_tr, yR_tr_log);
```

# 2. Optimization (Differential Evolution)

```python
# objective weights (empirically tuned)
wE, wM, wV, wR = 28.0, 32.0, 0.7, 1.4

# constraints for optimization
bounds = [(4.0, 5.0), (3.0, 5.0), (0.2, 5.0), (0.4, 1.0)]
penalty_weight = 1e6 

objective = lambda x: (
    -(
        wE * np.expm1(elon_model.predict(np.asarray(x, float).reshape(1,-1)).item()) -
        wM * np.expm1(mod_model.predict (np.asarray(x, float).reshape(1,-1)).item()) -
        wV * np.expm1(visc_model.predict(np.asarray(x, float).reshape(1,-1)).item()) -
        wR * np.expm1(rt_model.predict  (np.asarray(x, float).reshape(1,-1)).item())
     )
    + penalty_weight * (np.sum(np.asarray(x, float)) - 10.0)**2
)

result = differential_evolution(
    objective,
    bounds=bounds,
    maxiter=120,
    popsize=20,
    seed=42,
    disp=False,
    polish=True
)

x_best = result.x
print("Optimal parts (PUDA, IBOA, EHMA, MAA):", x_best)
```

# 3. Validation of Surrogate Models (K-fold CV, External validation set)

```python
TRAIN_DIR = "train_data"
VAL_DIR = "val_data"

X_val  = np.load(os.path.join(VAL_DIR, "X.npy")).astype(float)
yV_val = np.load(os.path.join(VAL_DIR, "y_viscosity.npy")).astype(float)
yE_val = np.load(os.path.join(VAL_DIR, "y_elongation.npy")).astype(float)
yM_val = np.load(os.path.join(VAL_DIR, "y_modulus.npy")).astype(float)
yR_val = np.load(os.path.join(VAL_DIR, "y_reactiontime2.npy")).astype(float)

# to fraction
rs_val = X_val.sum(axis=1)
if np.allclose(rs_val, 100.0, atol=1e-6):
    X_val = X_val / 10.0
elif np.allclose(rs_val, 1.0, atol=1e-6):
    X_val = X_val * 10.0

# stack all data
X  = np.vstack([X_tr, X_val])
yV = np.concatenate([yV_tr, yV_val])
yE = np.concatenate([yE_tr, yE_val])
yM = np.concatenate([yM_tr, yM_val])
yR = np.concatenate([yR_tr, yR_val])

# normalization
yV_log = np.log1p(yV)
yE_log = np.log1p(yE)
yM_log = np.log1p(yM)
yR_log = np.log1p(yR)

# K-fold CV
k = 5
kf = KFold(n_splits=k, shuffle=True, random_state=42)

oof_V = np.full_like(yV, np.nan, dtype=float)
oof_E = np.full_like(yE, np.nan, dtype=float)
oof_M = np.full_like(yM, np.nan, dtype=float)
oof_R = np.full_like(yR, np.nan, dtype=float)

V_metrics = []
E_metrics = []
M_metrics = []
R_metrics = []

for fold_idx, (tr_idx, te_idx) in enumerate(kf.split(X), start=1):
    X_train, X_test = X[tr_idx], X[te_idx]

    visc_model = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=0.01)
    elon_model = SVR(kernel='rbf', C=400, gamma=0.01, epsilon=0.01)
    mod_model  = SVR(kernel='rbf', C=17, gamma=0.04, epsilon=0.13)
    rt_model   = SVR(kernel='rbf', C=68, gamma=0.03, epsilon=0.1)

    visc_model.fit(X_train, yV_log[tr_idx])
    elon_model.fit(X_train, yE_log[tr_idx])
    mod_model.fit(X_train, yM_log[tr_idx])
    rt_model.fit(X_train, yR_log[tr_idx])

    yV_pred = np.maximum(np.expm1(visc_model.predict(X_test)), 0.0)
    yE_pred = np.maximum(np.expm1(elon_model.predict(X_test)), 0.0)
    yM_pred = np.maximum(np.expm1(mod_model.predict(X_test)), 0.0)
    yR_pred = np.maximum(np.expm1(rt_model.predict(X_test)), 0.0)

    oof_V[te_idx] = yV_pred
    oof_E[te_idx] = yE_pred
    oof_M[te_idx] = yM_pred
    oof_R[te_idx] = yR_pred

    V_r2_log = r2_score(np.log1p(yV[te_idx]), np.log1p(yV_pred))
    V_rmse_log = root_mean_squared_error(np.log1p(yV[te_idx]), np.log1p(yV_pred))
    V_r2_raw = r2_score(yV[te_idx], yV_pred)
    V_rmse_raw = root_mean_squared_error(yV[te_idx], yV_pred)
    V_metrics.append([V_r2_log, V_rmse_log, V_r2_raw, V_rmse_raw])

    E_r2_log = r2_score(np.log1p(yE[te_idx]), np.log1p(yE_pred))
    E_rmse_log = root_mean_squared_error(np.log1p(yE[te_idx]), np.log1p(yE_pred))
    E_r2_raw = r2_score(yE[te_idx], yE_pred)
    E_rmse_raw = root_mean_squared_error(yE[te_idx], yE_pred)
    E_metrics.append([E_r2_log, E_rmse_log, E_r2_raw, E_rmse_raw])

    M_r2_log = r2_score(np.log1p(yM[te_idx]), np.log1p(yM_pred))
    M_rmse_log = root_mean_squared_error(np.log1p(yM[te_idx]), np.log1p(yM_pred))
    M_r2_raw = r2_score(yM[te_idx], yM_pred)
    M_rmse_raw = root_mean_squared_error(yM[te_idx], yM_pred)
    M_metrics.append([M_r2_log, M_rmse_log, M_r2_raw, M_rmse_raw])

    R_r2_log = r2_score(np.log1p(yR[te_idx]), np.log1p(yR_pred))
    R_rmse_log = root_mean_squared_error(np.log1p(yR[te_idx]), np.log1p(yR_pred))
    R_r2_raw = r2_score(yR[te_idx], yR_pred)
    R_rmse_raw = root_mean_squared_error(yR[te_idx], yR_pred)
    R_metrics.append([R_r2_log, R_rmse_log, R_r2_raw, R_rmse_raw])

    print(
        f"Fold {fold_idx:2d} (n={len(te_idx):2d}) | "
        f"V: R2_log={V_r2_log:+.3f}, RMSE_log={V_rmse_log:.3f}, R2_raw={V_r2_raw:+.3f}, RMSE_raw={V_rmse_raw:.3f} || "
        f"E: R2_log={E_r2_log:+.3f}, RMSE_log={E_rmse_log:.3f}, R2_raw={E_r2_raw:+.3f}, RMSE_raw={E_rmse_raw:.3f} || "
        f"M: R2_log={M_r2_log:+.3f}, RMSE_log={M_rmse_log:.3f}, R2_raw={M_r2_raw:+.3f}, RMSE_raw={M_rmse_raw:.3f} || "
        f"R: R2_log={R_r2_log:+.3f}, RMSE_log={R_rmse_log:.3f}, R2_raw={R_r2_raw:+.3f}, RMSE_raw={R_rmse_raw:.3f}"
    )

V_metrics = np.array(V_metrics)
E_metrics = np.array(E_metrics)
M_metrics = np.array(M_metrics)
R_metrics = np.array(R_metrics)

# summary
print("\nCV summary (mean ± s.d. across folds)")
print(
    "Viscosity     : "
    f"R2_log = {V_metrics[:,0].mean():.4f} ± {V_metrics[:,0].std(ddof=1):.4f}, "
    f"RMSE_log = {V_metrics[:,1].mean():.4f} ± {V_metrics[:,1].std(ddof=1):.4f}, "
    f"R2_raw = {V_metrics[:,2].mean():.4f} ± {V_metrics[:,2].std(ddof=1):.4f}, "
    f"RMSE_raw = {V_metrics[:,3].mean():.4f} ± {V_metrics[:,3].std(ddof=1):.4f}"
)
print(
    "Elongation    : "
    f"R2_log = {E_metrics[:,0].mean():.4f} ± {E_metrics[:,0].std(ddof=1):.4f}, "
    f"RMSE_log = {E_metrics[:,1].mean():.4f} ± {E_metrics[:,1].std(ddof=1):.4f}, "
    f"R2_raw = {E_metrics[:,2].mean():.4f} ± {E_metrics[:,2].std(ddof=1):.4f}, "
    f"RMSE_raw = {E_metrics[:,3].mean():.4f} ± {E_metrics[:,3].std(ddof=1):.4f}"
)
print(
    "Modulus       : "
    f"R2_log = {M_metrics[:,0].mean():.4f} ± {M_metrics[:,0].std(ddof=1):.4f}, "
    f"RMSE_log = {M_metrics[:,1].mean():.4f} ± {M_metrics[:,1].std(ddof=1):.4f}, "
    f"R2_raw = {M_metrics[:,2].mean():.4f} ± {M_metrics[:,2].std(ddof=1):.4f}, "
    f"RMSE_raw = {M_metrics[:,3].mean():.4f} ± {M_metrics[:,3].std(ddof=1):.4f}"
)
print(
    "Reaction Time : "
    f"R2_log = {R_metrics[:,0].mean():.4f} ± {R_metrics[:,0].std(ddof=1):.4f}, "
    f"RMSE_log = {R_metrics[:,1].mean():.4f} ± {R_metrics[:,1].std(ddof=1):.4f}, "
    f"R2_raw = {R_metrics[:,2].mean():.4f} ± {R_metrics[:,2].std(ddof=1):.4f}, "
    f"RMSE_raw = {R_metrics[:,3].mean():.4f} ± {R_metrics[:,3].std(ddof=1):.4f}"
)

# OOF metrics on all 36 points
print("\nOOF metrics on all 36 points")
print(
    "Viscosity     : "
    f"R2_log = {r2_score(np.log1p(yV), np.log1p(oof_V)):.4f}, "
    f"RMSE_log = {root_mean_squared_error(np.log1p(yV), np.log1p(oof_V)):.4f}, "
    f"R2_raw = {r2_score(yV, oof_V):.4f}, "
    f"RMSE_raw = {root_mean_squared_error(yV, oof_V):.4f}"
)
print(
    "Elongation    : "
    f"R2_log = {r2_score(np.log1p(yE), np.log1p(oof_E)):.4f}, "
    f"RMSE_log = {root_mean_squared_error(np.log1p(yE), np.log1p(oof_E)):.4f}, "
    f"R2_raw = {r2_score(yE, oof_E):.4f}, "
    f"RMSE_raw = {root_mean_squared_error(yE, oof_E):.4f}"
)
print(
    "Modulus       : "
    f"R2_log = {r2_score(np.log1p(yM), np.log1p(oof_M)):.4f}, "
    f"RMSE_log = {root_mean_squared_error(np.log1p(yM), np.log1p(oof_M)):.4f}, "
    f"R2_raw = {r2_score(yM, oof_M):.4f}, "
    f"RMSE_raw = {root_mean_squared_error(yM, oof_M):.4f}"
)
print(
    "Reaction Time : "
    f"R2_log = {r2_score(np.log1p(yR), np.log1p(oof_R)):.4f}, "
    f"RMSE_log = {root_mean_squared_error(np.log1p(yR), np.log1p(oof_R)):.4f}, "
    f"R2_raw = {r2_score(yR, oof_R):.4f}, "
    f"RMSE_raw = {root_mean_squared_error(yR, oof_R):.4f}"
)
```

```python
EXT_VAL_DIR = "ext_val_data"

# load external validation
X_ext  = np.load(os.path.join(EXT_VAL_DIR, "X.npy")).astype(float)
yV_ext = np.load(os.path.join(EXT_VAL_DIR, "y_viscosity.npy")).astype(float)
yE_ext = np.load(os.path.join(EXT_VAL_DIR, "y_elongation.npy")).astype(float)
yM_ext = np.load(os.path.join(EXT_VAL_DIR, "y_modulus.npy")).astype(float)
yR_ext = np.load(os.path.join(EXT_VAL_DIR, "y_reactiontime2.npy")).astype(float)

# to fraction
rs_ext = X_ext.sum(axis=1)
if np.allclose(rs_ext, 100.0, atol=1e-6):
    X_ext = X_ext / 10.0
elif np.allclose(rs_ext, 1.0, atol=1e-6):
    X_ext = X_ext * 10.0

yV_hat_tr  = np.expm1(visc_model.predict(X_tr)).reshape(-1)
yE_hat_tr  = np.expm1(elon_model.predict(X_tr)).reshape(-1)
yM_hat_tr  = np.expm1(mod_model.predict(X_tr)).reshape(-1)
yR_hat_tr  = np.expm1(rt_model.predict(X_tr)).reshape(-1)

yV_hat_val = np.expm1(visc_model.predict(X_val)).reshape(-1)
yE_hat_val = np.expm1(elon_model.predict(X_val)).reshape(-1)
yM_hat_val = np.expm1(mod_model.predict(X_val)).reshape(-1)
yR_hat_val = np.expm1(rt_model.predict(X_val)).reshape(-1)

yV_hat_ext = np.expm1(visc_model.predict(X_ext)).reshape(-1)
yE_hat_ext = np.expm1(elon_model.predict(X_ext)).reshape(-1)
yM_hat_ext = np.expm1(mod_model.predict(X_ext)).reshape(-1)
yR_hat_ext = np.expm1(rt_model.predict(X_ext)).reshape(-1)

# combined validation metrics: original 3 + external 9
print("Validation metrics (original 3 + external 9) (raw R² | log RMSE)")
print("Viscosity     :", f"{r2_score(np.r_[yV_val, yV_ext], np.r_[yV_hat_val, yV_hat_ext]):.3f}", "|",
      f"{root_mean_squared_error(np.log1p(np.r_[yV_val, yV_ext]), np.log1p(np.r_[yV_hat_val, yV_hat_ext])):.3f}")
print("Elongation    :", f"{r2_score(np.r_[yE_val, yE_ext], np.r_[yE_hat_val, yE_hat_ext]):.3f}", "|",
      f"{root_mean_squared_error(np.log1p(np.r_[yE_val, yE_ext]), np.log1p(np.r_[yE_hat_val, yE_hat_ext])):.3f}")
print("Modulus       :", f"{r2_score(np.r_[yM_val, yM_ext], np.r_[yM_hat_val, yM_hat_ext]):.3f}", "|",
      f"{root_mean_squared_error(np.log1p(np.r_[yM_val, yM_ext]), np.log1p(np.r_[yM_hat_val, yM_hat_ext])):.3f}")
print("Reaction Time :", f"{r2_score(np.r_[yR_val, yR_ext], np.r_[yR_hat_val, yR_hat_ext]):.3f}", "|",
      f"{root_mean_squared_error(np.log1p(np.r_[yR_val, yR_ext]), np.log1p(np.r_[yR_hat_val, yR_hat_ext])):.3f}")

# parity plot
fig, axes = plt.subplots(2, 2, figsize=(9, 8))
pairs = [
    ("Viscosity",         yV_tr, yV_hat_tr, yV_val, yV_hat_val, yV_ext, yV_hat_ext, axes[0, 0]),
    ("Elongation",        yE_tr, yE_hat_tr, yE_val, yE_hat_val, yE_ext, yE_hat_ext, axes[0, 1]),
    ("Young's Modulus",   yM_tr, yM_hat_tr, yM_val, yM_hat_val, yM_ext, yM_hat_ext, axes[1, 0]),
    ("Reaction Time",     yR_tr, yR_hat_tr, yR_val, yR_hat_val, yR_ext, yR_hat_ext, axes[1, 1]),
]

for title, yt_tr, yp_tr, yt_val, yp_val, yt_ext, yp_ext, ax in pairs:
    lb = float(min(yt_tr.min(), yp_tr.min(), yt_val.min(), yp_val.min(), yt_ext.min(), yp_ext.min()))
    ub = float(max(yt_tr.max(), yp_tr.max(), yt_val.max(), yp_val.max(), yt_ext.max(), yp_ext.max()))

    ax.plot([lb, ub], [lb, ub], "k--", linewidth=1)
    ax.scatter(yt_tr,  yp_tr,  s=28, alpha=0.85, label="Train")
    ax.scatter(yt_val, yp_val, s=40, alpha=0.95, marker="s", color="orange",  label="Validation")
    ax.scatter(yt_ext, yp_ext, s=46, alpha=0.95, marker="D", color="crimson", label="External Validation")

    ax.set_title(title)
    ax.set_xlabel("Actual")
    ax.set_ylabel("Predicted")

axes[0, 0].legend()
plt.tight_layout()
plt.show()
```
